### **corpus 전처리**
- **1. 수자, 수자와 한글 혹은 영어가 같이 나오는 단어(3d, 6개, 60만 등등) 제거**
- **2. 한자는 제거, 한글과 영어만 추출**

```python
import re
from tqdm import tqdm

def preprocess_corpus(corpus):
    updated_corpus = []
    pattern_mixed = re.compile(r'(?=.*\d)(?=.*[a-zA-Z가-힣])')
    pattern_korean_english = re.compile(r'^[가-힣a-zA-Z]+$')

    for document in tqdm(corpus):
        if isinstance(document, str):
            words = document.split(',')
            filtered_words = [word for word in words if pattern_korean_english.match(word) and not pattern_mixed.search(word)]
            updated_corpus.append(','.join(filtered_words))
        else:
            updated_corpus.append('')
    return updated_corpus

cleaned_corpus = preprocess_corpus(corpus)
```

### **Corpus에서 제외하는 단어**
- **1. 단어출현빈도가 5개이하인 경우**

```python
from collections import Counter

# 단어의 출현 빈도를 계산합니다.
word_counts = Counter()

for document in tqdm(cleaned_corpus):
    words = document.split(',')
    for word in words:
        word_counts[word] += 1

# 출현 빈도가 5개 이하인 단어를 필터링합니다.
filtered_words = {word: count for word, count in word_counts.items() if count > 5}

# 필터링된 단어로 다시 corpus를 생성합니다.
filtered_corpus = []
for document in cleaned_corpus:
    words = document.split(',')
    filtered_document = [word for word in words if word in filtered_words]
    filtered_corpus.append(','.join(filtered_document))

# 필터링된 corpus 개수를 출력합니다.
print(len(filtered_corpus))
```

```python
# 샘플링: 예를 들어 전체 데이터의 10%만 사용
from sklearn.utils import resample
sampled_corpus = resample(filtered_corpus,
                          n_samples=int(0.1 * len(filtered_corpus)),
                          random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
import numpy as np

vect = CountVectorizer(stop_words = stop_words, ngram_range=(1, 1), min_df= 0.005, max_df= 0.8)
X = vect.fit_transform(sampled_corpus) # Doc-Term Matrix 계산
X = TfidfTransformer().fit_transform(X)

count = X.toarray().sum(axis=0) # Doc-Term Matrix 각 단어에 대한 합
idx = np.argsort(-count)
count = count[idx]

# Use get_feature_names_out() instead of get_feature_names()
feature_name = np.array(vect.get_feature_names_out())[idx]
keyword_list = list(zip(feature_name, count))
result_df = pd.DataFrame(keyword_list)
```
