### **비동기방식으로 수행**
#### API 호출을 100 나누어서 n번 수행하여 데이터프레임 합치기로

```python
import aiohttp
import asyncio
import nest_asyncio
import time

# Initialize nest_asyncio
nest_asyncio.apply()

start = time.time()

async def get_location(session, address, api_key):
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
    headers = {"Authorization": f"KakaoAK {api_key}"}
    try:
        async with session.get(url, headers=headers) as response:
            if response.status == 200:
                api_json = await response.json()
                if api_json['documents']:
                    x = api_json['documents'][0]['x']
                    y = api_json['documents'][0]['y']
                    return x, y
            print(f"Failed to get location for address: {address}, status: {response.status}")
    except Exception as e:
        print(f"Exception occurred for address: {address}, error: {e}")
    return None, None

async def fetch_coordinates(df_chunk, api_key):
    async with aiohttp.ClientSession() as session:
        tasks = []
        for idx, row in df_chunk.iterrows():
            address = row['시설도로명주소']
            tasks.append(get_location(session, address, api_key))
        coordinates = await asyncio.gather(*tasks)
    return coordinates

def process_chunk(df_chunk, api_key):
    coordinates = asyncio.run(fetch_coordinates(df_chunk, api_key))
    df_chunk['경도'], df_chunk['위도'] = zip(*coordinates)
    return df_chunk

def main(df, api_key):
    chunk_size = 100
    chunks = [df[i:i + chunk_size] for i in range(0, len(df), chunk_size)]

    processed_chunks = []
    for chunk in chunks:
        processed_chunk = process_chunk(chunk, api_key)
        processed_chunks.append(processed_chunk)

    result_df = pd.concat(processed_chunks, ignore_index=True)
    return result_df

# Run the main function and measure elapsed time
if __name__ == '__main__':
    updated_df = main(df, api_key)

    end = time.time()
    elapsed_time = (end - start) / 60
    print(f"총 걸린 시간(분) {elapsed_time:.2f} 분")

    # Display the DataFrame to check the added coordinates
    updated_df
```

### Folium 지도 그리기

```python
import folium

# Folium 맵 초기화
# 중심 좌표를 데이터프레임의 첫 번째 위도와 경도로 설정
center_lat = df['위도'].mean(); center_lon = df['경도'].mean()
map = folium.Map(location=[center_lat, center_lon], zoom_start=12)


# Add CircleMarkers to the map for each location
for idx, row in df.iterrows():
    if pd.notnull(row['위도']) and pd.notnull(row['경도']):
        folium.CircleMarker(
            location=[row['위도'], row['경도']],
            radius=3,  # Radius of the circle
            color='blue',  # Circle border color
            fill=True,
            fill_color='blue',
            fill_opacity=0.01  # Transparency of the circle
        ).add_to(map)

# Save the map to an HTML file
map.save('/content/drive/MyDrive/이대과제/excel/map.html')

# Display the map
map
```

### **비동기방식이나 API 한 번에 사용하려면 요금을 지불해야 해서 100개씩 나누어 위경도 추출**

```python
import requests
import aiohttp
import asyncio
import nest_asyncio
import time

# Initialize nest_asyncio
nest_asyncio.apply()

start = time.time()

async def get_location(session, address, api_key):
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
    headers = {"Authorization": f"KakaoAK {api_key}"}
    async with session.get(url, headers=headers) as response:
        if response.status == 200:
            api_json = await response.json()
            if api_json['documents']:
                x = api_json['documents'][0]['x']
                y = api_json['documents'][0]['y']
                return x, y
    return None, None

async def fetch_coordinates(df, api_key):
    async with aiohttp.ClientSession() as session:
        tasks = []
        for idx, row in df.iterrows():
            address = row['시설도로명주소']
            tasks.append(get_location(session, address, api_key))
        coordinates = await asyncio.gather(*tasks)
    return coordinates

def main(df, api_key):
    coordinates = asyncio.run(fetch_coordinates(df, api_key))

    # 결과를 데이터프레임에 추가
    df['경도'], df['위도'] = zip(*coordinates)

if __name__ == '__main__':
    main(df, api_key)

end = time.time()
elapsed_time = (end - start) / 60
print(f"총 걸린 시간(분) {elapsed_time:.2f} 분")
```
