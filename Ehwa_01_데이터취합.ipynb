```python
# 나눌 그룹 수
num_groups = 30

# 파일 리스트를 그룹으로 나누기
file_groups = [file_list[i::num_groups] for i in range(num_groups)]
```

## **엑셀파일 gzip파일로 일고 내 마운틴드라이브에 저장하기**

```python
import time
from tqdm import tqdm
import io
import gzip
import pickle
import pandas as pd
import os

# 코드 시작 시간 기록
start_time = time.time()

# chunk 파일을 읽을 나의 드라이브 디렉토리
folder_path = '/content/drive/MyDrive/이대과제/data1'
# chunks 디렉토리 경로, 디렉토리가 없으면 생성
os.makedirs(folder_path, exist_ok=True)

# 각 그룹에 대해 pickle 파일 생성
for i, group in enumerate(file_groups):
    # pickle 파일 이름
    gzip_filename = f"{folder_path}/articles_group_{i+1}.pkl.gz"

    #  메모리에 엑셀 파일을 다운로드하고 읽기
    articles = []
    for file in tqdm(group, desc=f'Processing group {i+1}/{num_groups}'):
        if file['mimeType'] == 'application/vnd.openxmlformats-officedocument.spreadsheetml.sheet':  # 엑셀 파일인 경우
            file_id = file['id']
            file_title = file['title']
            print(f"Reading {file_title}...")
            downloaded = drive.CreateFile({'id': file_id}) # Use the 'drive' object here

            # 파일 내용을 메모리 스트림으로 읽기
            downloaded.FetchContent()
            file_content = downloaded.content.getvalue()

            # 필요한 컬럼만 추출
            need_columns = ['일자', '언론사', '제목', '통합 분류1', '통합 분류2', '통합 분류3', '본문']
            # 메모리 스트림을 사용하여 데이터프레임으로 변환
            df = pd.read_excel(io.BytesIO(file_content), usecols=need_columns)
            display(df.head(2))

            articles.append(df)

    # 메모리에 있는 엑셀 파일을 gzip 압축된 pickle 파일로 저장
    with gzip.open(gzip_filename, 'wb') as f:
        pickle.dump(articles, f)

    print(f"Saved {gzip_filename}")

# 코드 종료 시간 기록
end_time = time.time()

# 소요된 전체 시간(분) 계산
execution_time = (end_time - start_time) / 60
print("코드 실행 시간:", execution_time, "분")
```

### 중복 디가 제거(일부) - 다시 제거해야함

```python
import pandas as pd
from tqdm import tqdm

file_path = '/content/drive/MyDrive/이대과제/data1/articles_group_28.pkl.gz'

for file_path in tqdm(pkl):
    with gzip.open(file_path, "rb") as f:
        # pickle 데이터 로드
        tf = pickle.load(f)
        dataframes = []
        ttl = []
        for i in range(len(tf)):
            ttl.append(len(tf[i]))
            dataframes.append(tf[i])
        df = pd.concat(dataframes)
        # 동일 기사 삭제
        df = df[~df.duplicated(['일자', '언론사', '제목'])]

    # 메모리에 있는 엑셀 파일을 pickle 파일로 저장
    with gzip.open(file_path, 'wb') as f:
        pickle.dump(df, f)
```

```python
# 모든 pickle 파일들을 읽어서 하나의 데이터프레임으로 합치기
# 기사의 제목이 동일한 contents를 제거
import pandas as pd
from tqdm import tqdm

dataframes = []
for pkl in tqdm(pkl_list):
    dfs_list = pd.read_pickle(pkl)
    print(f'파일 개수는 {len(dfs_list)}')
    for i in tqdm(range(len(dfs_list))):
        # 기사의 제목이 동일한 contents를 제거
        dfs_list[i] = dfs_list[i][~dfs_list[i].duplicated(subset=['제목'], keep='first')]
        print(f'제목 중복 제거된 파일 개수는 {dfs_list[i].shape}')
        dataframes.append(dfs_list[i])
    print('*'*100)

df = pd.concat(dataframes)
df.shape
```

### 각각 gzip파일을 다운로드하기

```python
from google.colab import files
import os

# chunks 디렉토리 경로
directory = 'chunks'

# 디렉토리 내 모든 파일에 대해 다운로드 수행
for filename in tqdm(os.listdir(directory)):
    filepath = os.path.join(directory, filename)
    if filename.endswith(".gz"):  # gzip 파일만 선택
        files.download(filepath)
```

```python
# prompt: /content/chunks/ 안의 모든 gzip파일을 하나의 데이터프레임으로 읽기

import time
# Start timing
start_time = time.time()

# chunks 디렉토리 경로
directory = '/content/chunks'

# 모든 gzip 파일 경로 목록 가져오기
gzip_files = [os.path.join(directory, filename) for filename in os.listdir(directory) if filename.endswith('.gz')]

# 데이터프레임 리스트 초기화
df_list = []

# 각 gzip 파일을 읽어서 데이터프레임으로 변환 후 리스트에 추가
for gzip_file in tqdm(gzip_files):
    with gzip.open(gzip_file, 'rb') as f:
        df_list.append(pickle.load(f))

# 데이터프레임 리스트를 하나의 데이터프레임으로 결합
df = pd.concat(df_list)

# End timing
end_time = time.time()
execution_time = end_time - start_time
print(f"Total execution time: {execution_time:.2f} seconds")
```
