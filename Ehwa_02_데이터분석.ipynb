### **재 시작**

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
%time df = pd.read_pickle('/content/drive/MyDrive/이대과제/df_gzip.pkl', compression='gzip')

# index 순차 넘버링
df.index = range(len(df))

# df['일자']를 날짜로 변경하고 index로 지정
df['일자'] = pd.to_datetime(df['일자'], format='%Y%m%d')
df.set_index('일자', inplace=True)

df
```

```python
dates = pd.date_range('2010', periods=15, freq='Y')
df.resample(rule='Y').count()['언론사'].plot(kind='bar', figsize=(15, 3), rot=30)
ax = plt.xticks(ticks=range(len(dates)), labels=[f'{year.year}년도' for year in dates])
plt.title(f'연도별 온라인기사 건 수', fontsize=16)
```

```python
df.columns
```

```python
df.reset_index(inplace=True)
df['년도']=df['일자'].apply(lambda x: x.year)
df['월']=df['일자'].apply(lambda x: x.month)
df = df[['일자', '언론사', '특성추출(가중치순 상위 50개)', '년도', '월']]
df
```

### 최근 5년동안 월별 컨텐츠 추이선

```python
import seaborn as sns
# Filter the DataFrame for years 2021, 2022, and 2023
df_filtered = df[df['년도'].isin([2019, 2020, 2021, 2022, 2023])]
# Group the filtered DataFrame by year and month and count the number of articles
grouped_df = df_filtered.groupby(['년도', '월']).size().unstack()
grouped_df.T
grouped_df.T.plot(figsize=(20, 3), rot=30)
plt.title(f'연도/월별 온라인기사 건 수(추이선)', fontsize=16)
```

```python
# 컨텐츠 게시 상위 30개 언론사
df.언론사.value_counts()[:30].plot(kind='bar', rot=45, figsize =(15,4)); plt.show()
```
